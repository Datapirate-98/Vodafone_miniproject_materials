# Q5. Find the state having current minimum Active patient and noof active
# patients.

```python
grouped_df1 = df1.groupby("State/UnionTerritory").sum()
grouped_df1
```

```python
d = dict(grouped_df1.idxmax())
print("The Maximun Number of active cases are in %s and no.of active patients = %.2f" %(d["Active"], max(grouped_df1["Active"])))
```

```python
d_min = dict(grouped_df1.idxmin())
print("The Minimun Number of active cases are in %s and no.of active patients = %.2f" %(d_min["Active"], min(grouped_df1["Active"])))
```

# Q6. Find the current Deaths to Cured ratio then find the states having
# minimum and maximum ratios.


```python
ratio = []
for death, cured in zip(grouped_df1["Deaths"], grouped_df1["Cured"]):
    try:
        ratio.append(death/cured)
    except:
        ratio.append(0)
ratio
```

```python
grouped_df1["death_cured_ratio"] = ratio
grouped_df1
```

```python
d_ratio = dict(grouped_df1.idxmax())
from fractions import Fraction
x = d_ratio["death_cured_ratio"]
value = max(grouped_df1["death_cured_ratio"])
frac = Fraction(value).limit_denominator()
print(f"The Maximun Deaths to Cured Ratio is at :{x} of ratio {frac}")
```

```python
dm_ratio = dict(grouped_df1.idxmin())
from fractions import Fraction
x = dm_ratio["death_cured_ratio"]
value = min(grouped_df1["death_cured_ratio"])
frac = Fraction(value).limit_denominator()
print(f"The Maximun Deaths to Cured Ratio is at :{x} of ratio {frac}")
```

# Q8.  Plot pie charts on the last recorded data of both Delhi and

# mention of percentage. Infer the meaning out of plots.

# **For Delhi**

```python
colors = ["green","red","blue"]
labels = ["Cured", "Active", "Deaths"]
c = delhi_df["Cured"].mean()
a = delhi_df["Active"].mean()
d = delhi_df["Deaths"].mean()
population = [c, a, d]
population
```
